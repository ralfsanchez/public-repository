```python
'''Instructions:

Make sure the uploaded notebook already displays output clearly 
because I will only be looking at GitHub (i.e., I will not run the notebook code myself).'''

#Use `requests` to track the location of the International Space Station. 

import requests
import time
import matplotlib.pyplot as plt

x_list = []
y_list = []

for x in range(100):
    Res = requests.get("http://api.open-notify.org/iss-now.json")
    obj = Res.json()

    #Add values in list
    x = obj['iss_position']['latitude']
    y = obj['iss_position']['longitude']

    x_list.append(x)
    y_list.append(x)
    
    
    fig, ax = plt.subplots()

    
    print("Timestamp for designated scatter plot below: " + str(obj['timestamp']))
    plt.scatter(x_list,y_list,label='ISS Station',color='k')
    plt.xlabel('latitude')
    plt.ylabel('longitude')
    plt.title(' Tracking the International Space Station')
    plt.legend()
    plt.show()
    time.sleep(10)




#Generate at least 100 data points with the latitude, longitude, and timestamp of the data point. You may do this by pinging this (Links to an external site.) API once every 10 seconds for 17 or so minutes.



#Use `matplotlib` to illustrate the path of the ISS. Use any visualization you feel is appropriate.

```
